## Introduction
The purpose of this project is to assess the lending risk for my client. A major source of revenue for my client is by charging interest on loans.
Lowering the risk of borrower defaults using the model, the lender will face a lower % of its borrowers defaulting and thus help the business to earn higher profits and maintain more customer relationships.

## Data:
The data is sourced from Kaggle (Lending Club) for the last three years (2017-2020). For this project, I'm using 246,000 data points with 28 features. My target for the project is Loan Status. I'll be analyzing two major categories in Loan Status: Charged off(Written as bad debt by the lender) and Fully Paid. Through Preliminary Analysis on the correlation between Employment length and Default status, borrowers with over 10 years of employment defaulted the most. Counter-intuitively, borrowers with less than 1 year of employment had a high payoff percentage.
